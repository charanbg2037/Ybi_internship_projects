##**Objective**

The objective of "Financial Market News Prediction" is to leverage predictive analytics and machine learning to forecast market movements, asset prices, and financial events based on the analysis of news and information relevant to the financial markets. This serves several key purposes:

Risk Management: To help investors, traders, and financial institutions anticipate and manage risks associated with market fluctuations, economic events, or news-driven volatility.

Regulatory Compliance: To help financial institutions comply with regulations related to market manipulation, insider trading, and information disclosure.

##**Import Library**

```python
import pandas as pd
import numpy as np
```

##**Import Data**

```python
data = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Financial%20Market%20News.csv', encoding = 'ISO-8859-1')
data
```

##**Describe Data**

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.columns
```

```python
' '.join(str(x) for x in data.iloc[0:, 2:27])
```

```python
news=[]
for rows in range(0, len(data.index)):
  news.append(' '.join(str(x) for x in data.iloc[rows, 2:27]))
```

```python
data.head()
```

```python
news[400]
```

```python
x = news
```

##**Define Target Variable (y) and Feature Variables (X):**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(lowercase = True, ngram_range=(1,1))
```

```python
x = cv.fit_transform(x)
```

```python
x.shape
```

```python
y = data['Label']
```

```python
y.shape
```

##**Train Test Split**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.3, random_state = 777)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

##**Modeling**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

##**Prediction**

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

##**Model Evaluation**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
acc = accuracy_score(y_test, y_pred)
print(acc)
```

```python
cun = confusion_matrix(y_test, y_pred)
print(cun)
```

```python
cal = classification_report(y_test, y_pred)
print(cal)
```
