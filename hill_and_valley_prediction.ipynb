#**HILL AND VALLEY PREDICTION**

##**Objective**

Predicting hills and valleys is a crucial objective in various fields:

1. Geology: To anticipate geological features like mountains and depressions for geological exploration and hazard assessment.

2. Agriculture: Identifying terrain variations to optimize crop planting, irrigation, and soil management for higher yields.

3. Urban Planning: For efficient infrastructure development and flood risk assessment in cities by mapping elevation changes.

4. Climate Science: Understanding local climate patterns through elevation data to predict weather and climate-related events.

5. Navigation: Aiding in route planning and GPS accuracy to ensure safe and efficient transportation.

6. Ecology: Identifying habitats and ecological niches within hills and valleys for conservation efforts.

7. Archaeology: Uncovering potential archaeological sites in valleys or elevated areas to discover historical artifacts and ancient settlements.

##**Define Target Variable (y) and Feature Variables (X):**

```python
y = data['Class']
print(y.shape)
x = data.drop(['Class'], axis = 1)
print(x.shape)
```

##**Data Preprocessing**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

```python
x.shape
```

##**Train Test Split**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size= 0.7, stratify= y, random_state= 30)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

##**Modeling**

```python
model =LogisticRegression()
model.fit(x_train, y_train)
```

```python
x_test
```

##**Prediction**

```python
y_pred = model.predict(x_test)
y_pred.shape
```

```python
y_pred
```

```python
model.predict_proba(x_test)
```

##**Model Evaluation**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(accuracy_score(y_pred, y_test))
```

```python
print(confusion_matrix(y_pred, y_test))
```

```python
print(classification_report(y_pred, y_test))
```
